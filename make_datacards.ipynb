```python
import json
def get_rate(json_fn, proc_name, region):
    with open(json_fn) as json_file:
        data = json.load(json_file)
        try:
            rate = data[region][proc_name]
        except:
            return 0
    return rate
```

```python
xsec_norm_str = {}
for k, v in xsec_norm_unc.items():
    unc_name = f"xsec_norm_{k}"
    xsec_norm_str[k] = f"{unc_name:<35}   lnN      "

bin_map = {'z_peak': 'z-peak','h_peak':'h-peak', 'h_sidebands':'h-sidebands'}
idx = []
rate = []
bin_names = []
proc_names = []

lumi_unc = 1.025
lumi_unc_list = []
all_bins = []
all_observations = []

for bin_name in bins:
    all_bins.append(bin_name)
    all_observations.append("-1")
    for i, p in enumerate(proc[bin_name]):
        if p in sig_proc:
            idx.append(f"{-i:<25}")
        else:
            idx.append(f"{i:<25}")
        proc_names.append(f"{p:<25}")
        the_rate = get_rate(rate_file, p, bin_map[bin_name])
        rate.append(f"{the_rate:<25}")
        bin_names.append(f"{bin_name:<25}")
        lumi_unc_list.append(f"{lumi_unc:<25}")
        for k, v in xsec_norm_unc.items():
            if p in k:
                xsec_norm_str[k] = xsec_norm_str[k]+f"{v:<25}" 
            else:
                dash = "-"
                xsec_norm_str[k] = xsec_norm_str[k]+f"{dash:<25}" 
                
    nproc = len(idx)
    nchannels = 1


all_bins_str = "  ".join(all_bins)
all_obs_str = "  ".join(all_observations)
bin_str = "".join(bin_names)
proc_str = "".join(proc_names)
idx_str = "".join(idx)
rate_str = "".join(rate)
lumi_unc_str = "".join(lumi_unc_list)

syst_strings = {}
syst_sources = {
    'muSF': proc,
    'pu_weight': proc,
}
if '2018' not in year:
    syst_sources.update{'l1prefiring_weight': proc,}

for s, sp in syst_sources.items():
    syst_strings[s] = f"{s:<35}   shape    "
    for b in bins:
        for p in proc[b]:
            if p in sp[b]:
                syst_strings[s] += f"{'1.0':<25}"
            else:
                syst_strings[s] += f"{'0.0':<25}"
            
datacard_name = f'combine/datacard_test_{label}_{suff}_{year}.txt'
```
